```python
CCSN_df = read_CCSNe_hosts_data()
CCSN_df = CCSN_df[CCSN_df["z"]<=0.2].reset_index()
CCSN_df.head(2)
```

```python
logM = np.linspace(5, 12, 100)

counters1 = []
for k in tqdm(range(1)):
    counter = np.zeros(np.shape(logM))
    for i in (range(1000)):
        ccsn = CCSN_df.sample(8).logM # sample size = 8 (dsa only)
        for j in range(len(logM)):
            if sum(ccsn<logM[j]) == 0:
                counter[j] += 1
    counter /= 1000
    counters1.append(counter)


counters2 = []
for k in tqdm(range(1)):
    counter = np.zeros(np.shape(logM))
    for i in (range(1000)):
        ccsn = CCSN_df.sample(20).logM # sample size = 20 (all FRBs)
        for j in range(len(logM)):
            if sum(ccsn<logM[j]) == 0:
                counter[j] += 1
    counter /= 1000
    counters2.append(counter)


counters3 = []
for k in tqdm(range(1)):
    counter = np.zeros(np.shape(logM))
    for i in (range(1000)):
        ccsn = CCSN_df.sample(12).logM # sample size = 12 (gordon+chime only)
        for j in range(len(logM)):
            if sum(ccsn<logM[j]) == 0:
                counter[j] += 1
    counter /= 1000
    counters3.append(counter)
```

```python
import scipy
f = scipy.interpolate.interp1d(logM, counters1[0])
print(f(9))

f = scipy.interpolate.interp1d(logM, counters2[0])
print(f(9))

f = scipy.interpolate.interp1d(logM, counters3[0])
print(f(9))
```
